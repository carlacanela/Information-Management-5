```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_path = './rpi_describe.csv'

SF = pd.read_csv(dataset_path)
```

```python
SF.head()
```

```python
SF.describe
```

```python
data = SF #['weight']
data['rounded'] = round(data['weight'])

data['diff'] = 20 - data['rounded'] 
data.head()
```

```python
data['rounded'].describe()
```

```python
count = data['rounded'].count()
mean = data['rounded'].mean() #gemiddelde
median = data['rounded'].median()
std = data['rounded'].std()
# rng = data['rounded'].rng()
```

```python
print ("Count = {}".format(count))  
print ("Mean = {}".format(mean))  
print ("Median = {}".format(median))  
print ("Std = {:.2}".format(std))  
```

```python
freq = data['rounded'] #.value_counts()
print(freq)

freq = (freq).to_frame()
print(freq)


freq = freq.reset_index()
print(freq)

print(type(freq))
```

```python
freq.columns = ['value', 'freq']
print(freq)
```

```python
plt.figure(figsize=(20, 10))

plt.ylabel('Frequency')
plt.xlabel('Weight')

plt.plot(freq.value, freq.freq, "o", markersize=10, color='g')
```
