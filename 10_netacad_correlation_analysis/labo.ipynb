```python
import pandas as pd 
file = './brainsize.txt'
DF = pd.read_csv(file)
```

```python
DF.head()
```

```python
DF.describe()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

men = DF[DF['Gender'] == 'Male']
female = DF[DF['Gender'] == 'Female']
```

```python
men_avg_smarts = men[["PIQ", "FSIQ", 'VIQ']].mean(axis=1)
plt.scatter(men_avg_smarts, men['MRI_Count'])
plt.show()
```

```python
female_avg_smarts = female[["PIQ", "FSIQ", 'VIQ']].mean(axis=1)
plt.scatter(female_avg_smarts, men['MRI_Count'])
plt.show()
```
